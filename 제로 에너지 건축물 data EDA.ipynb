### 에너지 자립률 분석

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#한글폰트
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 한글 폰트 경로 설정
font_path = 'C:/Windows/Fonts/malgun.ttf'  # 여기서 'malgun.ttf'는 예시로 사용한 것이며, 실제 한글 폰트 파일로 변경해야 합니다.
font_name = fm.FontProperties(fname=font_path).get_name()
plt.rc('font', family=font_name)

# 에너지 자립률 칼럼을 숫자형으로 변환
data['에너지자립률'] = pd.to_numeric(data['에너지자립률'], errors='coerce')

# 에너지 자립률을 10 단위로 그룹화하는 함수
def group_energy_autonomy_rate(rate):
    return (rate // 10) * 10

data['에너지 자립률 그룹'] = data['에너지자립률'].apply(group_energy_autonomy_rate)

# 에너지 자립률 그룹별 박스플롯
plt.figure(figsize=(12, 6))
sns.boxplot(x='에너지 자립률 그룹', y='에너지자립률', data=data, palette='viridis')
plt.title('에너지 자립률 그룹별 박스플롯')
plt.xlabel('에너지 자립률 그룹')
plt.ylabel('에너지 자립률')
plt.show()
```

```python
# 에너지 자립률이 250 이상인 건물 삭제
data = data[data['에너지자립률'] < 250]
```

```python
# 지역별 에너지 자립률 그룹의 박스플롯
plt.figure(figsize=(14, 8))
sns.boxplot(x='지역', y='에너지 자립률 그룹', data=data, palette='viridis')
plt.title('지역별 에너지 자립률 그룹의 박스플롯')
plt.xlabel('지역')
plt.ylabel('에너지 자립률 그룹')
plt.xticks(rotation=45)
plt.show()
```

```python
# ZEB 등급별 에너지 자립률 그룹의 박스플롯
plt.figure(figsize=(14, 8))
sns.boxplot(x='제로에너지건축물 인증등급(ZEB)', y='에너지 자립률 그룹', data=data, palette='viridis')
plt.title('제로에너지건축물 인증등급(ZEB)별 에너지 자립률 그룹의 박스플롯')
plt.xlabel('ZEB 인증등급')
plt.ylabel('에너지 자립률 그룹')
plt.xticks(rotation=45)
plt.show()
```

```python
# 건물주용도와 지역의 조합별 건물 수 계산
building_use_region_counts = data.groupby(['건물주용도', '지역']).size().unstack().fillna(0)

# 도넛 차트 그리기 함수
def plot_donut(data, title):
    plt.figure(figsize=(8, 8))
    plt.pie(data, labels=data.index, autopct='%1.1f%%', startangle=140, pctdistance=0.85)
    centre_circle = plt.Circle((0, 0), 0.70, fc='white')
    fig = plt.gcf()
    fig.gca().add_artist(centre_circle)
    plt.title(title)
    plt.axis('equal')
    plt.show()

# 각 건물주용도별 지역의 도넛 차트 그리기
for use in building_use_region_counts.index:
    plot_donut(building_use_region_counts.loc[use], f'{use}의 지역별 건물 수 분포')
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.decomposition import PCA
import seaborn as sns
import matplotlib.pyplot as plt

# 데이터 불러오기
df = pd.read_csv('C:/Users/82106/Desktop/데이터 분석 프로젝트 2/데이콘 빅데이터 분석/data/한국에너지공단_제로에너지건축물 인증현황_20231231.csv',encoding='cp949')

#한글폰트
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 한글 폰트 경로 설정
font_path = 'C:/Windows/Fonts/malgun.ttf'  # 여기서 'malgun.ttf'는 예시로 사용한 것이며, 실제 한글 폰트 파일로 변경해야 합니다.
font_name = fm.FontProperties(fname=font_path).get_name()
plt.rc('font', family=font_name)
```

```python
# 모든 값을 문자열로 변환 후 공백 제거
df = df.applymap(lambda x: str(x).strip())

# 에너지자립률 열에서 실수형으로 변환할 수 없는 값 제거
df = df[df['에너지자립률'].apply(lambda x: x.replace('.', '', 1).isdigit())]
df['에너지자립률'] = df['에너지자립률'].astype(float)

# 에너지자립률이 200 이상인 데이터 삭제
df = df[df['에너지자립률'] < 200]

# 인증일자 열에서 날짜 형식으로 변환할 수 없는 값 제거
df = df[pd.to_datetime(df['인증일자'], errors='coerce').notna()]
df['인증일자'] = pd.to_datetime(df['인증일자'])

# 카테고리형 데이터 라벨 인코딩
label_encoder = LabelEncoder()
categorical_columns = ['인증구분', '건물주용도', '건물용도', '지역', '건축물명', '업체명', '제로에너지건축물 인증등급(ZEB)']

for col in categorical_columns:
    df[col] = label_encoder.fit_transform(df[col])
```

```python
# 2. 상관관계 분석
correlation_matrix = df.corr()

# 상관 행렬 출력
print("Correlation Matrix:")
print(correlation_matrix)

# 3. 상관 행렬 히트맵 시각화
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix Heatmap')
plt.show()
```
